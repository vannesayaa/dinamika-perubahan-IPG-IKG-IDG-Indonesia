# Dinamika Perubahan IKG, IDG, IPG 6 Tahun Terakhir 

```python
# mengambil data indonesia setiap ikg, idg, dan ipg 
ipg = df_ipg.loc [df_ipg['Kota dan Provinsi'] == 'INDONESIA']
idg = df_idg.loc [df_idg['Kota dan Provinsi'] == "INDONESIA"]
ikg = df_ikg.loc [df_ikg['Kota dan Provinsi'] == "INDONESIA"]
```

```python
# transformasi kolom menjadi baris 
ipg = ipg.melt ()
idg = idg.melt () 
ikg = ikg.melt ()
```

```python
# mengambil baris yang digunakan
ipg = ipg[1:]
idg = idg[1:]
ikg = ikg[1:]
```

```python
# mengganti tipe data menjadi numeric
ipg = ipg.astype ('float')
idg = idg.astype ('float')
ikg = ikg.astype ('float')
```

```python
# mengurutkan data berdasarkan tahun terkecil ke tahun yang lebih besar 2018 menuju 2023
ipg = ipg.sort_values ('variable', ascending = True)
idg = idg.sort_values ('variable', ascending = True)
ikg = ikg.sort_values ('variable', ascending = True)
```

```python
ikg
```

```python
#standarisasi data 
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler ()

# mengambil 
nilai_ipg = ipg.iloc [:, 1:]
nilai_idg = idg.iloc [:, 1:]
nilai_ikg = ikg.iloc [:, 1:]
nilai1 = pd.DataFrame (scaler.fit_transform (nilai_ipg), columns = nilai_ipg.columns)
nilai2 = pd.DataFrame (scaler.fit_transform (nilai_idg), columns = nilai_idg.columns)
nilai3 = pd.DataFrame (scaler.fit_transform (nilai_ikg), columns = nilai_ikg.columns)
```

```python
nilai1
```

```python
import matplotlib.pyplot as plt
# Penggabungan menjadi satu diagram 
plt.plot (nilai1, color = 'red', label = "Indeks Pembangunan Gender ")
plt.plot (nilai2, color = 'green', label = "IDG")
plt.plot (nilai3, color = 'violet', label = "Indeks Kesetaraan Gender")
plt.xlabel ('Tahun')
plt.ylabel ('Nilai')
plt.title ('Diagram Garis IPG, IDG, dan IKG')
plt.legend ()
plt.show()
```
